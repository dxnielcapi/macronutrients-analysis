# Data Macronutrients
Put parameters to separate all the values, we need to separate all the aliments in three important groups
1. Proteins
2. Carbohydrates
3. Lipids


```python
#Steps to this part
# - Separate the values in groups
# - Transform the values to have the correct information

#Drop values to have a dataset only with macronutrients
drop_values = ['Data.Alpha Carotene', 'Data.Ash', 'Data.Beta Carotene',
                'Data.Beta Cryptoxanthin', 'Data.Cholesterol', 'Data.Choline',
                'Data.Fiber', 'Data.Lutein and Zeaxanthin', 'Data.Lycopene',
                'Data.Manganese', 'Data.Niacin', 'Data.Pantothenic Acid',
                'Data.Refuse Percentage', 'Data.Retinol', 'Data.Riboflavin',
                'Data.Selenium', 'Data.Thiamin', 'Data.Major Minerals.Calcium',
                'Data.Major Minerals.Copper', 'Data.Major Minerals.Iron', 'Data.Major Minerals.Magnesium',
                'Data.Major Minerals.Phosphorus', 'Data.Major Minerals.Potassium', 'Data.Major Minerals.Sodium',
                'Data.Major Minerals.Zinc', 'Data.Vitamins.Vitamin A - IU', 'Data.Vitamins.Vitamin A - RAE',
                'Data.Vitamins.Vitamin B12', 'Data.Vitamins.Vitamin B6', 'Data.Vitamins.Vitamin C',
                'Data.Vitamins.Vitamin E', 'Data.Vitamins.Vitamin K'
              ]
```

```python
#Create a new dataframe with the vital information of the macronutrients
#    Variables 
#    data_macronutrients: Dataset with the iformation of the three macronutrients

data_macronutrients = data_food.drop(columns = drop_values, axis = 1, inplace=False)
print(data_macronutrients.columns)
```

```python
#The data of the actual dataset have a distinct weights and this cause problems to know all the values in each parameter.
#Function to create the criteria for a standar of 100g 

def criteria(data_macronutirents, parameter):
    
    # Use both columns to obtain the value if one is missing | If one is NaN or 0 go to the second parameter
    data_macronutrients["Weight"] = np.where(
        (data_macronutrients["Data.Household Weights.1st Household Weight"].isna()) |
        (data_macronutrients["Data.Household Weights.1st Household Weight"] == 0),
        data_macronutrients["Data.Household Weights.2nd Household Weight"],
        data_macronutrients["Data.Household Weights.1st Household Weight"]
    )

    # Avoid divisions by 0
    data_macronutrients["Weight"].replace(0, np.nan, inplace=True)

    # Create a new column for the update data (100 gr)
    data_macronutrients[f"{parameter} (100g)"] = (data_macronutrients[parameter] / data_macronutrients["Weight"]) * 100

    # Retorn the new column 
    return data_macronutrients
```

```python
#data_macronutrients = criteria(data_macronutrients, "Data.Fat.Total Lipid")
# Variables 
#    macro_data_updated: Updated values in a criteria of 100gr 

#Calling the function using a for statement to transform many columns

parameters = [
       'Data.Carbohydrate', 'Data.Protein',
       'Data.Fat.Monosaturated Fat',
       'Data.Fat.Polysaturated Fat', 'Data.Fat.Saturated Fat',
       'Data.Fat.Total Lipid', 'Data.Kilocalories',
       'Data.Sugar Total', 'Data.Water',
]

for param in parameters:
    macro_data_updated = criteria(data_macronutrients, param)
```

```python
macronutrients_data_updatedd = macronutrients_data_updated.dropna()
```

```python
# Filtering all the data in three groups: Proteins, Carbohydrates, Lipids 
# All the parameters are based in the documentation


#The function introduce three parameters two compare important things about the macronutrients

def filter_nutrients(macronutrients_data_updatedd, 
                     first_colm, first_parameter, 
                     second_colm, second_parameter, 
                     third_colm, third_parameter):
    return macronutrients_data_updatedd[
        (macronutrients_data_updatedd[first_colm] > first_parameter) &
        (macronutrients_data_updatedd[second_colm] < second_parameter) &
        (macronutrients_data_updatedd[third_colm] < third_parameter)
    ]
```

```python
data_lipids=filter_nutrients(
    macronutrients_data_updatedd,
    first_colm = "Data.Fat.Total Lipid (100g)",
    first_parameter = 15,
    second_colm = 'Data.Kilocalories (100g)',
    second_parameter = 300,
    third_colm = 'Data.Fat.Saturated Fat (100g)',
    third_parameter = 18
)
```
